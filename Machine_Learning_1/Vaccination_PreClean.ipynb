```python
import pandas as pd
import os
```

```python
file_path = "../Cleaned_Data/final_project_data.csv"
data = pd.read_csv(file_path)
data
```

```python
data.columns
```

```python
# Drop unneccessary columns
dropped_data = data.drop(columns = ['country_name-2','iso_code-2','Indicator Name','Indicator Code','2020'])
```

```python
# Only keep rows where 2019 GDP is no NA
clean_data = dropped_data[dropped_data['2019'].notna()]
clean_data
```

```python
# Unique dates 
clean_data.date.nunique()
```

```python
# Unique Countries
clean_data.country_name.nunique()
```

```python
# What dates do we have 
dates = clean_data['date'].unique()
dates.sort()
print(dates)
```

```python
# Find best date (the one that returns the most countries) to filter our data 
country_count_df = clean_data.groupby('date')['country_name'].nunique()

pd.options.display.max_rows = 76

country_count_df
```

```python
# Filter data by 2021-03-18
regression_data = clean_data[clean_data['date']=='2021-03-18']
regression_data = regression_data.rename(columns={'2019': 'GDP'})
regression_data
```

## Linear Regression

```python
from pathlib import Path
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
ordered_data = regression_data[['country_name', 'iso_code', 'GDP', 'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'daily_vaccinations_raw', 'daily_vaccinations', 'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred']]
ordered_data.sort_values(by='total_vaccinations', ascending=False)
```

```python
plt.scatter(ordered_data.GDP, ordered_data.total_vaccinations_per_hundred)
plt.xlabel('GDP in $US Dollars')
plt.ylabel('Total Vaccinations Given per 100 People')
plt.show()
```

```python
ordered_data.to_csv('GDP_Vaccincation_Clean.csv')
```

## Linear Regression

```python
# Format the independent variable to meet Skikit-learn's requirements
X = ordered_data['GDP'].values.reshape(-1,1)
```

```python
print(X.shape)
```

```python
# Set Dependent Variable
y = ordered_data['total_vaccinations_per_hundred']
```

```python
# Instantiate linear regression model
model = LinearRegression()
```

```python
# Train the model
model.fit(X,y)
```

```python
# Create predictions
y_pred = model.predict(X)
```

```python
# Plot the data and the best fit line
plt.scatter(X,y)
plt.plot(X, y_pred, c='red')
plt.xlabel("GDP in $US Dollars")
plt.ylabel("Total Vaccinations per 100 People")
plt.show()
```
