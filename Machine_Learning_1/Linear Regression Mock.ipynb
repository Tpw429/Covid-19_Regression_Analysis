```python
# Load the data file
final_data = pd.read_csv('country_master_join_vaccination_table.csv')
final_data
```

```python
#Get the datatype 
final_data.dtypes
```

```python
# Get the count of rows and columns for both DataFrames
final_data.shape
```

```python
# Format the independent variable to meet Scikit-learn's requirements
X = final_data['total_vaccinations'].values.reshape(-1, 1)
```

```python
print(X.shape)
```

```python
# The dependent variable
y = final_data['people_fully_vaccinated']
```

```python
# Instantiate linear regression model
model = LinearRegression()
```

```python
# Train the model
model.fit(X, y)
```

```python
# Create predictions
y_pred = model.predict(X)
```

```python
# Plot the data and the best fit line
plt.scatter(X, y)
plt.plot(X, y_pred, c='red')
plt.xlabel("total_vaccinations")
plt.ylabel("People fully vaccinated")
plt.show()
```
